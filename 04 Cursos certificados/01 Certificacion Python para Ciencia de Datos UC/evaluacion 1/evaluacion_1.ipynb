```python
#4.Pregunta 4: ¿Cuál cifra corresponde al promedio diario de pacientes que han llegado a la red en los últimos
# 10 días (desde el día 415 al día 424)?  
info_hist = pd.read_excel("datos.xls", sheet_name='Información Histórica')
info_hist
```

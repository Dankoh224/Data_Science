# Clustering con el MÃ©todo K-Means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Importamos los datos

```python
vinos = pd.read_csv('vinito.csv')
vinos.info()
```

```python
vinos.head()
```

```python
# Quitamos primera columna:
vinos_variables = vinos.drop(['Vino'], axis = 1)
```

```python
# Obtenemos estadisticos de vinos
vinos_variables.describe()
```

```python
# Normalizamos para convertir los valores para que se encuentren en un mismo rango
vinos_norm = (vinos_variables-vinos_variables.min())/(vinos_variables.max()-vinos_variables.min())
vinos_norm
```

```python
vinos_norm.describe()
```

  # Busqueda de la cantidad optima de clusters, suponiendo que no tenemos un criterio para decidir cuantos cluesters a crear y que ampoco sabemos cual serÃ­a la cantidad Ã³ptima de clusters a formar.

## Se usa el mÃ©todo llamado codo de jambÃº. Que consiste en ir creando diferentes cantidades de clusters y calculando que tan similares son los individuos dentro de los mismose ir plasmando esta infromaciÃ³n en una grÃ¡fica.

## Mientras mÃ¡s similares son estos, mÃ¡s distantes se encontraran los clusters que se formen que es lo que buscamos en el clutering, la formaciÃ³n de grupos ien delimitados que contengn individuos cuya distancia entre ellos sea la menor posible.

## Para medir esta distancia se utiliza la suma de los cuadrados de cada grupo tambiÃ©n llamada WCSS, por sus siglas en inglÃ©s Within-Cluster-Sum-of-Squares.

#### Calculando que tan similares son los individuos dentro de los cluesters

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, max_iter = 300)
    kmeans.fit(vinos_norm) # AquÃ­ anota que aplica K-means a la base de datos
    wcss.append(kmeans.inertia_)
```

# Graficando los resultados de WCSS para formar el Codo de JambÃº y obtener la cantidad de Clusters necesarios.

```python
plt.plot(range(1,11),wcss)
plt.title('Codo de JambÃº')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('WCSS') # Es un indicador de quÃ© tan similares son los individuos dentro de los clusters
plt.show()
```

### Se observa en el grÃ¡fico que a medida que avanzan los nÃºmeros de clusters, disminuyen los WCSS. TambiÃ©n se puede notar que dejan de decrecer de forma drÃ¡stica en nÃºmero de clusters igual a 3, para decrecer lentamente. Esto significa que solo necesitamos 3 clusters para no tener demasiadas particiones.

# Aplicando el modelo k-means a la base de datos 

```python
clustering = KMeans(n_clusters = 3, max_iter = 300)
clustering.fit(vinos_norm)
```

# Agregando la clasificaciÃ³n al archivo original.

```python
vinos['KMeans_Clusters'] = clustering.labels_ # Los resultados del clustering se guardan en labels_ dentro del modelo
vinos.head()
```

# VisualizaciÃ³n de los clusters 
### Aplicaremos el anÃ¡lisis de componentes principales (PCA) para darnos una unda idea de cÃ³mo se formaron los clusters.

```python
 from sklearn.decomposition import PCA
    
pca = PCA(n_components=2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df = pd.DataFrame(data = pca_vinos, columns = ['Componente_1', 'Componente_2'])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[['KMeans_Clusters']]], axis = 1)

pca_nombres_vinos
```

```python
# Coloreamos los clusters

fig = plt.figure(figsize = (6,6)) #creamos una figura de 6x6

ax = fig.add_subplot(1,1,1) #indicamos que haremos solo un grÃ¡fico dentro de la figura
ax.set_xlabel('Componente 1', fontsize = 15) #nombres del eje x
ax.set_ylabel('Componente 2', fontsize = 15) #nombre del eje y
ax.set_title('Componentes Principales', fontsize = 20) #nombre del grÃ¡fico

color_theme = np.array(['blue', 'green', 'orange']) #creamos un array con tres colores para cada cluster
ax.scatter(x = pca_nombres_vinos.Componente_1, y = pca_nombres_vinos.Componente_2,
          c = color_theme[pca_nombres_vinos.KMeans_Clusters], s = 50)
#dibujar los puntos que representarÃ¡n los vinos con scatter
#agregamos los valores de las columnas componentes
#usamos el parÃ¡metro c para indicarle los colores que deseamos para los puntos
# esta informaciÃ³n se encuentra en KMeans_Clusters del objeto pca_nombres_vinos
# el tamaÃ±o de los puntos a graficar serÃ¡ de 50
plt.show()
```

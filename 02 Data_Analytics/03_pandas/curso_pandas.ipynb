# Series

#### Son estructuras de datos que contienen dos vectores; los indices y los datos, esto quiere decir que solo es una columna.

```python
meses = pd.Series(['enero','febrero','marzo','abril','mayo'], name = "Meses")
meses
```

#### NOTA: la serie puede tener nombre pero no tendrá etiqueta de columna, ya que solo es un columna de archivo (ni idea que quiere decir esto)

## Indice ALFANUMERICO

```python
serie_alfanumerica = pd.Series(['enero','febrero','marzo','abril','mayo'], index = ['a','b','c','d','e'])
serie_alfanumerica
```

## Indice con números distintos a los predefinidos

```python
serie_distinto_indice = pd.Series(['enero','febrero','marzo','abril','mayo'], index = ['1','2','3','4','5'])
serie_distinto_indice
```

## Indice con fechas

```python
fechas = pd.date_range('20200101', periods = 5)
datos = ['enero','febrero','marzo','abril','mayo']
serie_fechas = pd.Series(datos, index = fechas)
serie_fechas
```

# DataFrames

#### Almacenar datos de pasajeros del Titanic. 

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
            "Valderrama, Mr Danko"
        ],
        "Age": [22, 35, 58,36],
        "Sex": ["male", "male", "female","male"],
        "Higth": [1.76,1.77,1.78,1.79],
    }
)
df
```

```python
df['Age']
```

## Máximos de un dataframe o serie

```python
# maximo de la columna edad creada anteriormente
df['Age'].max()
```

```python
# Serie nueva:
Ages = pd.Series([22, 35, 58], name="Ages")
Ages.max()
```

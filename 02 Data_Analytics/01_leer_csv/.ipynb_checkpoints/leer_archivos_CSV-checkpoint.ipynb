### Mostrar filas de de un punto en adelante (header = el valor que estará en la cabecera será el anterior a esa posición).

### Ordenar tabla de menor a mayor, según una columna en particular

```python
datos.sort_values(by = 'Teléfono')
```

### Ordenar tabla de mayor a menor, según una columna en particular

```python
datos.sort_values(by = 'Teléfono', ascending = False)
```

### Mostrar los datos menores a 92 de las primeras 3 filas

```python
datos[datos[0:3]<92]
```

#### Muestra quienes cumplen o no la condición

```python
datos['Teléfono'] < 10
```

#### Muestra solo los elementos que cumplen la condición en una columna

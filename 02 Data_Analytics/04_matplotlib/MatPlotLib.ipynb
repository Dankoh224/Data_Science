```python
import pandas as pd
```

```python
# crear lista de rrss
fbk = ["Facebook",2449,True,2006]
twt = ["Twitter",339,False,2006]
ig = ["Instagram",1000,True,2010]
yt = ["Youtube",2000,False,2005]
lkn = ["LinkedIn",663,False,2003]
wsp = ["Whatsapp",1600,True,2009]
```

```python
# Crear lista con las listas anteriores
lista_rrss = [fbk,twt,ig,yt,lkn,wsp]
```

```python
# Crear un DataFrame
df_rrss = pd.DataFrame(lista_rrss, columns = ["Nombre","Cantidad","ES_FBK","Año"])
df_rrss
```

```python
# Un ejemplo de un gráfico de lineas supermegabásico es:
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

```python
# Ahora realizaremos el grafico de LINEAS con NOMBRE y CANTIDAD:
plt.plot(df_rrss["Nombre"],df_rrss["Cantidad"])
plt.show()
```

```python
# Ahora realizaremos el grafico de PUNTOS con NOMBRE y CANTIDAD:
plt.scatter(df_rrss["Nombre"],df_rrss["Cantidad"])
plt.show()
```

```python
# Se puede realizar lo mismo que en la celda anterior con un método de pandas:
df_rrss.plot(kind = "scatter", x = "Nombre", y = "Cantidad")
plt.show()
```

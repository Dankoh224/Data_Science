# Imputar datos faltantes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('AbandonoEmpleados.csv', sep = ';', index_col= 'id', na_values='#N/D')
df
```

```python
df.isnull().sum().sort_values(ascending = False)
```

# Eliminar todas las filas que tienen datos faltantes

```python
df = df.dropna()
df
```

## Ahora veremos como nos quedó

```python
df.isnull().sum().sort_values(ascending = False)
```

# IMPUTAR DATOS FALTANTES

```python
df = pd.read_csv('AbandonoEmpleados.csv', sep = ';', index_col= 'id', na_values='#N/D')
df
```

```python
# Usaremos la función describe, donde podemos ver todos los datos y observar la media por columna para reemplazarla
# en los valores nulos
df['sexo'].describe()
```

```python
# Y para reemplazar
df['anos_en_puesto'] = df['anos_en_puesto'].replace(np.nan,2.77)

df['sexo'] = df['sexo'].replace(np.nan,2.72)

df.isnull().sum().sort_values(ascending = False)
```

# Otra forma de imputar los valores es usando Sklearn

```python
df = pd.read_csv('insurance.csv')
df
```

```python
# Exportaré un excel para eliminar algunos datos para que hayan nulos
df.to_excel("insurance.xlsx", sheet_name="passengers", index=False)
```

```python
df1 = pd.read_excel('insurance.xlsx')
df1
```

```python
df1 = pd.get_dummies(df1)
df1
```

```python
df1.isnull().sum().sort_values(ascending = False)
```

```python
# Importaremos sklearn para entrenar un modelo que rellene todas
# las variables, continuas o categóricas
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp.fit(df1)
df1 = pd.DataFrame(imp.transform(df1), columns = ['age','bmi','children','charges','sex_female','sex_male','smoker_no','smoker_yes','region_northeast','region_northwest','region_southeast','region_southwest'], dtype = int)
df1
```

```python
df1.isnull().sum().sort_values(ascending = False)
```
